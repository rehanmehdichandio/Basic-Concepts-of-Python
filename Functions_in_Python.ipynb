```python
def multiply(*args):
  product = 1
  for i in args:
    product = product * i
  print(args)
  print(product)
```

```python
multiply(1,2,3,4)
```

```python
#kwargs

def display(**kwargs):
  for (key, value) in kwargs.items():
    print(key, '-->', value)

display(India = 'Delhi', Pakistan = 'Islamabad')
```

**Lambda Function**


*   No Name
*   Lambda has no return value (infact, returns a function)
*   Lambda is written in 1 line
*   Not Reusable

Then why use lambda functions?
They are used with HOF (Higher Order Function)





```python
 #check if a string has 'a'

x = lambda s: 'a' in s
x("Hello")
```

```python
#odd or even
y= lambda a : 'even' if a%2==0 else 'odd'

y(4)
```
